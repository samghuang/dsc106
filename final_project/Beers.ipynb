```python
beers = pd.read_csv('beers.csv')
beers
```

```python
breweries = pd.read_csv('breweries.csv')
breweries = breweries.rename(columns={"Unnamed: 0": "brewery_id"})
breweries.head()
```

```python
# merge beers and breweries 
beers_breweries = beers.merge(breweries, left_on='brewery_id', right_on='brewery_id')
beers_breweries = beers_breweries.drop(columns=['Unnamed: 0'])
beers_breweries = beers_breweries.rename(columns={"name_x": "beer", "name_y":"brewery"})
beers_breweries.head()
```

### Simplify style 

```python
def simply_type(style):
    if isinstance(style, str):   
        if 'Lager' in style:
            return 'Lager'
        elif 'APA' in style:
            return 'APA'
        elif 'IPA' in style:
            return 'IPA'
        elif 'Stout' in style:
            return 'Stout'
        elif 'Porter' in style:
            return 'Porter'
        elif 'Ale' in style:
            return 'Ale'
        elif 'Cider' in style:
            return 'Cider'
        elif 'Pilsner' in style:
            return 'Pilsner'
        else: 
            return 'Other'
    else:
        return 'Other'

beers_breweries['style_simplified'] = beers_breweries['style'].apply(simply_type)
```
